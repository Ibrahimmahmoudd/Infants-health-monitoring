```python
import pandas as pd
```

```python
df = pd.read_csv(r"../data/newborn_health_monitoring_with_risk.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Heart Rate
df = df[(df["heart_rate_bpm"] > 50) & (df["heart_rate_bpm"] < 250)]

# Respiratory Rate
df = df[(df["respiratory_rate_bpm"] > 10) & (df["respiratory_rate_bpm"] < 100)]

# Oxygen Saturation
df = df[(df["oxygen_saturation"] >= 50) & (df["oxygen_saturation"] <= 100)]

# Temperature
df = df[(df["temperature_c"] > 30) & (df["temperature_c"] < 45)]
```

```python
df.describe()
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
```
